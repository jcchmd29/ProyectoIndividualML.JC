Descarga descomprimido json items como DataFrame

```python
def lac():
    with open(ruta_items, 'rt', encoding='utf-8') as file:
        return [ast.literal_eval(line.strip()) for line in file]     

datos = lac()
df_user_items = pd.DataFrame(datos)
```

Desanidar los valores dentro de la columna items

```python

# Función para extraer datos de la columna "items"
def extract_item_data(items):
    if not items:
        return pd.Series({'item_id': None, 'item_name': None, 'playtime_forever': None})
    item = items[0]
    return pd.Series({'item_id': item['item_id'], 'item_name': item['item_name'], 'playtime_forever': item['playtime_forever']})


# Aplicar la función para crear nuevas columnas
new_columns = df_items_items['items'].apply(extract_item_data)

# Combinar las columnas extraídas con el DataFrame original
df_items_items_playtime = pd.concat([df_items_items, new_columns], axis=1)

# Eliminar la columna original "items" y "item_id"
df_items_items_playtime.drop('items', axis=1, inplace=True)
df_items_items_playtime.drop('item_id', axis=1, inplace=True)

# Ver el DataFrame resultante
df_items_items_playtime
```

```python
# Convertir la columna 'playtime_forever' a tipo entero (int)
df_items_items_playtime['playtime_forever'] = df_items_items_playtime['playtime_forever'].fillna(0).astype(float)

# Crear una nueva columna con 'playtime_forever' transformada en horas
df_items_items_playtime['playtime_hours'] = (df_items_items_playtime['playtime_forever'] / 3600)
```

```python
df_items_items_playtime['playtime_hours'] = df_items_items_playtime['playtime_hours'].round(2)
```
