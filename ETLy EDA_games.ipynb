```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
import ipywidgets as widgets 
import ast
import gzip
import pandas as pd
from io import StringIO
```

Crear rutas de archivos

```python
# ruta de archivos JSON

ruta_games = "PI MLOps - STEAM/output_steam_games2.json"
```

Leer archivo games json

```python
# leer archivo games

# Crea una lista para almacenar los objetos JSON
lista_de_json = []

# Abre el archivo JSON y lee línea por línea
with open(ruta_games, 'r') as archivo_json:
    for linea in archivo_json:
        objeto_json = json.loads(linea)
        lista_de_json.append(objeto_json)
```

Transformar a un DF

```python
# Crea un DataFrame a partir de la lista de objetos JSON
df_games = pd.DataFrame(lista_de_json)
```

```python
df_games.sample(3)
```

Eliminar espacios en blanco

```python
# Aplicar strip() a cada celda del DataFrame para eliminar espacios en blanco
df_games = df_games.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

Crear DF con columnas a utilizar

```python
df_games_geners_app = df_games[['genres', 'app_name','user_id']]
df_games_geners_app
```

Observacion: La columna genres tiene datos anidados tipo lista

Extraer los valores tipo lista de la columna genres a valores indiviuales

```python

# la función str.join para combinar los elementos de la columna 'genres' en una cadena separada por comas.
df_games_geners_app['genres'] = df_games_geners_app['genres'].apply(lambda x: ', '.join(x) if isinstance(x, list) else x)

# Se divide la columna 'genres' en columnas separadas para cada género utilizando str.get_dummies.
df_genres_split = df_games_geners_app['genres'].str.get_dummies(', ')

# Concatenar las columnas generadas al DataFrame original.
df_games_genres_app = pd.concat([df_games_geners_app, df_genres_split], axis=1)
```

```python
# mostrar el DF
df_games_geners_app.sample(10)
```

Ahora la columna genres posee datos separados por coma de tipo str

Vamos crear una columna por cada valor en la columna genres

```python

# Itera a través de las columnas generadas a partir de 'genres' y crea una nueva columna para cada valor dividido.
for genre_column in df_genres_split.columns:
    # Crea una nueva columna con el nombre del género y asigna 1 si ese género está presente en esa fila, 0 en caso contrario.
    df_games_genres_app[genre_column] = df_genres_split[genre_column]


df_games_genres_app.sample(10)
```

```python
# Veamos las columnas del DF para excluir las que no utilizaremos
df_games_genres_app.columns
```

Eliminar columnas que no voy a utilizar, me quedo con las que voy a trabajar

```python

columnas_a_mantener = ['user_id','app_name','Action', 'Adventure', 'Racing', 'Simulation', 'Strategy','Sports']

# Crear una lista de columnas que deseas mantener
columnas_a_eliminar = [col for col in df_games_genres_app.columns if col not in columnas_a_mantener]

# Eliminar las columnas no deseadas del DataFrame
df_games_genres_app = df_games_genres_app.drop(columnas_a_eliminar, axis=1)
```

```python
# Crear una serie con los datos de la columna 'user_id' para rellenar los valores NaN

valores_aleat = df_games_geners_app['user_id'][df_games_geners_app['user_id'].notna()]
type(valores_aleat)
```

```python
valores_aleat
```

```python
# crear una lista con los nombres de app_name para rellenar los NaN de esa misma columna
app_name_aleat = df_games_geners_app['app_name'][df_games_geners_app['app_name'].notna()]
type(app_name_aleat)
```

```python
app_name_aleat
```

```python

# Crear una copia del DataFrame 
df_copy = df_games_genres_app.copy()
df_copy
```

```python
# funcion para rellenar los Null de la columna

def llenar_user_id(row):
    if pd.notna(row['app_name']) and pd.isna(row['user_id']):
        return np.random.choice(valores_aleat)
    else:
        return row['user_id']
```

```python

# Aplica la función para llenar los valores NaN en 'user_id'
df_copy['user_id'] = df_games_geners_app.apply(llenar_user_id, axis=1)
```

```python

# Llena los valores nulos en la columna 'app_name' con valores aleatorios de la serie
df_copy['app_name'] = df_copy['app_name'].fillna(app_name_aleat, inplace=True)
```

```python
# Converti columnas a tipo numérico

numericas=['Action', 'Adventure', 'Racing', 'Simulation', 'Strategy','Sports']

# Convierte las columnas especificadas a tipo numérico
df_copy[numericas] = df_copy[numericas].apply(pd.to_numeric, errors='coerce')


```
